Let's open over dataset "potus_visitors_2015.csv".

```python
dataset=list(reader(open("potus_visitors_2015.csv",encoding="utf-8")))
header=dataset[0] # saving headers of dataset into header.
dataset=dataset[1:]# saving data without header into dataset.
```

Now we retrieves the third element (index 2) of the first row of the dataset and assigns it to the variable x. It then retrieves the third element of the last row of the dataset and assigns it to the variable y. Both x and y is string data type.

```python
x=dataset[0][2]
y=dataset[-1][2]
print(x,y)
```

```python
Method='%m/%d/%Y %H:%M'
```

In above statement, Method='%m/%d/%Y %H:%M', %m, %d, %Y, %H, and %M are format codes that represent the month (01-12), day (01-31), year (e.g. 2022), hour (00-23), and minute (00-59), respectively.

```python
str_time=dt.datetime.strptime(x,Method)
print(str_time)
```

Above statement converts a string x to a datetime object using the strptime method.

```python
str_time.month # accessing the month component of the datetime using the month attribute.
```

```python
str_time.hour # accessing the hour component of the datetime using the hour attribute.
```

```python
str_time.minute # accessing the minute component of the datetime using the minute attribute.
```

```python
time_str=dt.datetime.strftime(str_time,Method)
print(time_str)
```

Above statement converts a datetime object str_time to a formatted string using the strftime method.

In our data set of potus visitors we have a column of start date at index(2) and End date at index(3). The datatype of this column is string. By applying foor loop let's convert it into date time using strptime.

```python
for i in dataset:
    i[2]=dt.datetime.strptime(i[2],Method)
    i[3]=dt.datetime.strptime(i[3],Method)
    print(i[2],i[3])
```

Now creating the frequency table of number of visitors in each month.

```python
visitors_count={}
for i in dataset: 
    month=i[2].strftime("%B")
    if month in visitors_count:
        visitors_count[month]+=1
    else:
        visitors_count[month]=1
```

Here's a step-by-step explanation of the provided code that creates a frequency table of the number of potus visitors in each month:

1. Initialize an empty dictionary named visitors_count to store the frequency counts.

2. Iterate over each item i in the dataset.

3. Extract the month from the third element of each item i using the strftime("%B") method. This converts the datetime object in i[2] to a string representation of the month's full name.

4. Check if the extracted month is already a key in the visitors_count dictionary.

5. If the month is already a key, increment the count associated with that month by 1 by using the += operator.

6. If the month is not a key, add it to the visitors_count dictionary with an initial count of 1.

7. Repeat steps 2-6 for all items in the dataset.

8. After executing this code, the visitors_count dictionary will contain the frequency counts of potus visitors in each month. 

```python
visitors_count # Each key-value pair in the dictionary will represent a month and its corresponding count.
```

### Time Delta
    timedelta is a class in the datetime module in Python that represents a duration or difference between two datetime objects. It can be used to perform arithmetic operations with dates and times.
    For Example,

```python
current_time=dt.datetime.now() #now () is a method of datetime which creat the object representing current date and time.
current_time
```

```python
thirty_days = dt.timedelta(days=30)
```

```python
new_date=current_time+thirty_days
print(new_date)
```

In above three line, the code first retrieves the current date and time using datetime.now(). It then creates a timedelta object representing a duration of 30 days. By adding this duration to the current time using the + operator, a new datetime object is calculated, representing the date and time that is 30 days in the future. Finally, the new date and time are printed. This demonstrates how datetime and timedelta can be used together to perform calculations involving dates and times in Python.

# Functions

In Python, a function is a named block of code that performs a specific task. Functions are a fundamental building block of programming, as they allow you to break complex tasks down into smaller, more manageable pieces of code that can be reused across your program.

Here's an overview of the different parts of a Python function:

    def function_name(parameter1, parameter2, ...):

        "Docstring - description of the function."
    
        # Code to perform the task
        return result

1. def: This keyword is used to define a function in Python.
2. function_name: This is the name of the function. Function names should be descriptive and follow the same naming conventions as variable names (e.g. lowercase with underscores between words).
3. (parameter1, parameter2, ...): These are the function's parameters, which are optional inputs that can be passed to the function when it is called. Multiple parameters can be separated by commas.
4. Docstring: This is an optional multi-line string at the beginning of a function that describes the function's purpose, inputs, and outputs. It is good practice to include a docstring for every function you write.
5. Code to perform the task: This is the code block that performs the specific task of the function. This block can contain any valid Python code, including other functions and control structures like loops and conditionals.
6. return result: This statement returns the output of the function to the calling code. If the function does not return anything, the return statement can be omitted.

    Here's an example of a simple function that adds two numbers together and returns the result:

```python
def add_numbers(num1, num2):
    "Add two numbers together and returns the result."
    result = num1 + num2
    return result
```

```python
add_numbers(8,9)
```

```python
def Multiply_numbers(num1, num2):
    "Multiply two numbers together and returns the result."
    return num1 * num2
    
```

```python
Multiply_numbers(6,4)
```

## Types of Parameter
### 1. Positional Parameter
    These are the most common type of function parameter. When you call a function, the arguments are assigned to the parameters in the order they are declared in the function definition.

```python
def operation1(a,b,c):
    return a+b,a-c
```

```python
operation1(8,9,7) # 8,9,7 arguments assigned to positional parameters a,b,c
```

### 2. Keyword Parameters:
    Keyword parameters are used to specify arguments by name rather than by position. This can be useful when calling functions with many arguments or when you want to skip optional arguments. For example:

```python
def operation2(a,b,c):
    return a+b,a-c
```

```python
operation2(8,c=7,b=9)
```

### 3. Default Parameters:
    Default parameters are used to specify a default value for a parameter. If an argument is not provided for a default parameter, the default value is used instead. For example:

```python
def operation3(a=8,b=9,c=7):
    return a+b,a-c
```

```python
operation3()
```

### Function of Frequency Table

Let's import the data set of apple store and create a function of frequency table using parameter "dataset" which is the original apple store data and "col_num" representing column number.

```python
apple_store=list(reader(open("Applestore.csv",encoding="utf-8")))
```

```python
def frq_table1(dataset,col_num): 
    dictionary={}
    for i in dataset[:1]:
        header =i[col_num]
    for i in dataset[1:]:
        if i[col_num] in dictionary:
            dictionary[i[col_num]]+=1
        else:
            dictionary[i[col_num]]=1
    return header, dictionary 
```

This function frq_table(dataset,col_num) takes a dataset (list of lists) and a column number (col_num) as input, and returns a frequency table as a dictionary. Here's a breakdown of how the function works:

* The function initializes an empty dictionary called dictionary.
* It loops through the first row of the dataset and assigns the header of the column specified by col_num to the variable header.
* It then loops through the rest of the dataset (excluding the first row).
* For each row, it checks if the value of the column specified by col_num is already a key in the dictionary.
* If it is, the function increments the value associated with that key by 1.
* If it is not, the function adds the value as a new key in the dictionary and sets its value to 1.
* Once the loop is finished, the function returns a tuple containing the header and dictionary.

```python
frq_table1(apple_store,11) # Frequency Table of prime genre using data set and column number
```

Now we will make the same function but instead of using column number (col_num) as second parameter we will use column name (col_name) as second parameter.

```python
def frq_table2(dataset,col_name):
    dictionary={}
    index=dataset[0].index(col_name)
    for i in dataset[:1]:
        header =col_name
    for i in dataset[1:]:
        if i[index] in dictionary:
            dictionary[i[index]]+=1
        else:
            dictionary[i[index]]=1
    return header, dictionary 
```

This is a modified version of the frq_table() function that takes a dataset and a column name as input instead of a column number. Here's a breakdown of how this function works:

* The function initializes an empty dictionary called dictionary.
* It finds the index of the column specified by col_name in the first row of the dataset and assigns it to the variable index.
* It sets the header variable to col_name.
* It loops through the rest of the dataset (excluding the first row).
* For each row, it checks if the value of the column specified by col_name (found using the index variable) is already a key in the dictionary.
* If it is, the function increments the value associated with that key by 1.
* If it is not, the function adds the value as a new key in the dictionary and sets its value to 1.
* Once the loop is finished, the function returns a tuple containing the header and dictionary.

```python
frq_table2(apple_store,"prime_genre")# Frequency Table of prime genre using data set and column name.
```
