```python
def is_english(app_name):
    lst = []
    for i in app_name:
        if ord(i) > 127:
            lst.append(False)
        else:
            lst.append(True)
    
    if False in lst:
        return False
    else:
        return True
```

```python
def is_english(app_name):
    lst = []
    for i in app_name:
        if ord(i) > 127:
            lst.append(False)
        else:
            lst.append(True)
            
    non_ascii = 0
    
    for j in lst:
        if j == False:
            non_ascii += 1
    
    if non_ascii > 3:
        return False
    else:
        return True
```

```python
is_english("Instagram😋😋😋😋")
```

# Most Common Genres as per avg Installation

```python
avg_installation = []

for cat in uni_cat:
    total_int = 0
    no_ins = 0
    
    for i in andriod_final:
        categories = i[1]
        if cat == categories:
            n_install = i[5]
            n_install = int(n_install.replace("+","").replace(",",""))
            total_int += n_install
            no_ins += 1
            
    avg_ins = total_int/no_ins
    avg_installation.append([round(avg_ins),cat])
```

```python
sorted(avg_installation, reverse = True)
```
