```python
import pandas as pd
```

```python
df = pd.read_csv('Football 17-18.csv')
df.head()
```

```python
team_table = pd.DataFrame(columns=['Team', 'Matches Played', 'Points', 'Wins', 'Losses', 'Draws', 'Goal For', 'Goal Against', 'Goal Difference'])
```

```python
grouped = df.groupby(['HomeTeam', 'AwayTeam'])
```

```python
for group_key, group_df in grouped:
    print("Group Key:", group_key)
    print(group_df)
    print("\n")
```

```python
for (home_team, away_team), group in grouped:
    home_goals = group['FTHG'].sum()
    away_goals = group['FTAG'].sum()
    result = group['FTR'].values[0]
    
    if not home_team in team_table['Team'].values:
        team_table = team_table.append({'Team': home_team, 'Matches Played': 0, 'Points': 0, 'Wins': 0, 'Losses': 0, 'Draws': 0, 'Goal For': 0, 'Goal Against': 0, 'Goal Difference': 0}, ignore_index=True)
    if not away_team in team_table['Team'].values:
        team_table = team_table.append({'Team': away_team, 'Matches Played': 0, 'Points': 0, 'Wins': 0, 'Losses': 0, 'Draws': 0, 'Goal For': 0, 'Goal Against': 0, 'Goal Difference': 0}, ignore_index=True)
    
    team_table.loc[team_table['Team'] == home_team, 'Matches Played'] += 1
    team_table.loc[team_table['Team'] == home_team, 'Goal For'] += home_goals
    team_table.loc[team_table['Team'] == home_team, 'Goal Against'] += away_goals
    
    team_table.loc[team_table['Team'] == away_team, 'Matches Played'] += 1
    team_table.loc[team_table['Team'] == away_team, 'Goal For'] += away_goals
    team_table.loc[team_table['Team'] == away_team, 'Goal Against'] += home_goals
    
    if result == 'H':
        team_table.loc[team_table['Team'] == home_team, 'Points'] += 3
        team_table.loc[team_table['Team'] == home_team, 'Wins'] += 1
        team_table.loc[team_table['Team'] == away_team, 'Losses'] += 1
    elif result == 'A':
        team_table.loc[team_table['Team'] == away_team, 'Points'] += 3
        team_table.loc[team_table['Team'] == away_team, 'Wins'] += 1
        team_table.loc[team_table['Team'] == home_team, 'Losses'] += 1
    else:
        team_table.loc[team_table['Team'] == home_team, 'Points'] += 1
        team_table.loc[team_table['Team'] == away_team, 'Points'] += 1
        team_table.loc[team_table['Team'] == home_team, 'Draws'] += 1
        team_table.loc[team_table['Team'] == away_team, 'Draws'] += 1

```

```python
team_table['Goal Difference'] = team_table['Goal For'] - team_table['Goal Against']
```

```python
team_table = team_table.sort_values(by='Points', ascending=False).reset_index(drop=True)
```

```python
print("Premier League Final Standing Table:")
print(team_table)
```

```python
team_table.to_csv('PremierLeagueStanding1.csv', index=False)
```
