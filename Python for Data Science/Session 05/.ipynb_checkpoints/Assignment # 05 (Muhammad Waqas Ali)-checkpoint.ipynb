### String.replace()
    The replace() method is a built-in function in Python that is used to replace one or more occurrences of a substring in a string with another substring. The syntax of the replace() method is as follows:
    string.replace(old, new).replace(old, new)...
    For Example,

```python
"(American)".replace("(","").replace(")","")
# Replacing "(" and ")" with empty string "".
```

In our dataset at index[2], nationality of the artist is listed in an odd format like (Nationality). So we have to remove the open bracket and close bracket to make it formally formated.

```python
for i in artwork:
    i[2]=i[2].replace("(","").replace(")","")
    print(i[2])
```

Similar problem with gender column at index[5] and will perform the same operation at index[5].

```python
for i in artwork:
    i[-3]=i[-3].replace("(","").replace(")","")
    print(i[-3])
```

### Begin and End Date

In column number 4 and 5 of our data set we have born year (BeginDate) and death year(EndDate) started with "-" which is not formaticaly correct so we have to remove the "-" from the begining of each year.

```python
# Removing "-" from begin date.
for i in artwork:
    if i[3]!="":
        i[3]=i[3].replace("-","")
        print(i[3])
```

```python
# Removing "-" from end date.
for i in artwork:
    if i[4]!="":
        i[4]=i[4].replace("-","")
        print(i[4])
```

### Date (Submission Year)

In column number 7 we have Date with several unwanter character in it. We have to clean it in order to get the desired result. Firt we make a list of possible characters that can be present in our Date column.

```python
bad_Char=["(",")","s","S","C","c","'","."," "]
print(bad_Char)
```

For example we have a string "c. 1998's" and we have to remove the unwanted characters from this string to make it 1998. We use for loop.

```python
s = "c. 1998's"
for i in bad_Char:
    s = s.replace(i, "")

print(s)
```

Here we can see "1998" without any unwanter character. What happen in the above code, we iterate over each character in the list(bad_Char) using a loop. Inside the loop, we use the replace() method to replace each occurrence of the current character with an empty string.

Let's apply the same algorithm in date column of our dataset.

```python
for i in artwork:
    for j in bad_Char:
        i[-2] = i[-2].replace(j,"")
    print(i[-2])
```

Here's a step-by-step breakdown of the above code:

1. The bad_Char list contains a set of characters that we want to remove from the date string.
2. The outer loop iterates over each element in the artwork list.
3. The inner loop iterates over each character in the bad_Char list.
4. For each element in artwork, the inner loop replaces all occurrences of the current bad_Char character with an empty string in the second-to-last element (i.e. the date string). This is done using the replace() method of the string.
5. Finally, the cleaned date string is printed for each element in artwork.

### Removing "-" 
In our date column along with other characters we also have "-" and we can not remove it in above for loop because "-" is placed in two different loaction first at the begining of date and 2nd in between the years which basically defining the range of year first we remove the "-" from the begining of date/year. 
He we use another method of string startswith()


#### Startswith()
The startswith() method is a built-in Python string method that returns True if a string starts with a specified substring and False otherwise. Let's apply this in our dataset to remove "-" from the begining of date column.

```python
for i in artwork:
    if i[-2].startswith("-"):
        i[-2]=i[-2].replace("-","")
    print(i[-2])
        
    
```

Now in above data we have years of length 4 and range of years length 9. Let's check if we have some other item of diffent length.

```python
for i in artwork:
    if "-" in i[-2] and (len(i[-2]) != 9 and len(i[-2]) != 4):
        print(i[-2])
```

Well we can see over here data some year range have missing century. So we have make it correct. Let's assume that the start and end of century is same.

```python
for i in artwork:
    if "-" in i[-2] and len(i[-2]) != 9:
        value = i[-2]
        value = value[:5] + value[:2] + value[5:]
        i[-2] = value
        print(i[-2])
```

Here's a step-by-step explanation of the code:

1. Loop through each item in the artwork list using a for loop.

2. Check if the current item's i[-2] string contains a hyphen (-) and is not of length 9. This condition is checked using the if statement with the help of the in and len functions.

3. If the condition is True, then initialize the variable value to the current item's i[-2] string.

3. Then, reformat the value string by concatenating the first five characters with the first two characters and the remaining characters of the string.

4. Assign the reformatted value string to the current item's i[-2] position.

5. Finally, print the i[-2] value for the current item.

The purpose of this code is to reformat the year range string if it is not in the correct format. The correct format is a string with a length of 9 and a hyphen in the fifth position (e.g. "1910-30"). The code checks for this condition and, if not met, reformats the string to the correct format by swapping the first five and the first two characters.







Now we only have year and range of year. Well from our dataset we can see that it is the submission year and range of year is inappropriate from. So we will take the average of range of year using split method of string.

```python
for i in artwork:
    if "-" in i[-2]:
        year=i[-2].split("-")
        avg=round((int(year[0])+int(year[1]))/len(year))
        i[-2]=avg
        print(year,i[-2])
```

Here is a step-by-step explanation of the code:

1. The code iterates over each element i in the artwork.
2. Then if statement checks whether the string i[-2] contains a dash ("-").
3. If a dash is found, the string is split into two parts based on the dash using the split() method, and the two parts are assigned to a list called year.
4. The code calculates the average of the two years using the formula avg=round((int(year[0])+int(year[1]))/len(year)). Here, int() is used to convert the two-year strings to integers before addition and round() is used to round the average to the nearest integer. The resulting value is assigned to i[-2].

```python
for i in artwork:
    i[-2]=int(i[-2])
    print(i[-2])
```

Now we can see that we have a clean date(subission of year column).


## Actual Age
    Now we will add the column of actual age to list down the age of artist.
    First append the actual age in the header.

```python
header.append("Actual Age")
header
```

Now we find the age by running the below code.

```python
for i in artwork:
    if i[3]!="" and i[4]!="":
        i[3] = int(i[3])
        i[4] = int(i[4])
        age = i[4] - i[3]
        i.append(age)  
    else:
        i.append("-")
```

Here are the step-by-step explanations for the given code:

1. The for loop iterates over each element in the artwork.
2. Then if statement checks if the i[3] and i[4] are empty or not.
3. If they are not empty, the code converts i[3] and i[4] from strings to integers using the int() function.
4. The difference between i[4] and i[3] is calculated and stored in a variable age.
5. age is appended as a new element to the current element (i.append(age)).
6. If either i[3] or i[4] is empty, a dash (-) is appended as a new element to the current element (i.append("-")).

```python
print(artwork[:21])
```

We can see that age column is added.

## Submission Age
    Now we will add the column of submission age to list down the submission age of artist.
    First append the submission age in the header.

```python
header.append("Submission Age")
header
```

Now we find the submission age by running the below code.

```python
for i in artwork:
    if i[3]!="" and i[6]!="":
        i[3] = int(i[3])
        i[6] = int(i[6])
        sub_age = i[6] - i[3]
        i.append(sub_age)  
    else:
        i.append("-")
```

It first checks if the element at index 3 and index 6 are not empty strings ("") using the logical operator and.

If they are not empty strings, it converts the elements at index 3 and index 6 to integers using the int() function. It then calculates the difference between the two integers and stores it in a variable sub_age.

Finally, it appends the value of sub_age to the end of the i list using the append() method.

If either element at index 3 or index 6 is an empty string, it appends a hyphen (-) to the end of the i list instead.

## Write into CSV file
Now after cleaning of data we will make a new csv file to save our clean and formated data. For this we will use "writer" method from csv. Below is the code to write the cleaned data in csv file.

```python
from csv import writer
with open('New American Gallery.csv', mode='w', encoding='utf-8', newline='') as file:
    writer = writer(file)
    writer.writerows(artwork)
```

Here is the explanation of above code.

1. from csv import writer: Import the writer class from the csv module.
2. with open('New American Gallery.csv', mode='w', encoding='utf-8', newline='') as file: Open a new file called "New American Gallery.csv" in write mode with UTF-8 encoding and without adding extra newlines between rows. The with statement ensures that the file is properly closed when the block of code is finished.
3. writer = writer(file): Create a writer object by passing the open file handle to the writer() function.
4. writer.writerows(artwork): Write the contents of the artwork list to the CSV file using the writerows() method of the writer object.
